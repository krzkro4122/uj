# Preliminary analysis 

```python
dataset_filename = 'phishing-websites.arff'
columns = [
    "having_IP_Address",
    "URL_Length",
    "Shortining_Service",
    "having_At_Symbol",
    "double_slash_redirecting",
    "Prefix_Suffix",
    "having_Sub_Domain",
    "SSLfinal_State",
    "Domain_registeration_length",
    "Favicon",
    "port",
    "HTTPS_token",
    "Request_URL",
    "URL_of_Anchor",
    "Links_in_tags",
    "SFH",
    "Submitting_to_email",
    "Abnormal_URL",
    "Redirect",
    "on_mouseover",
    "RightClick",
    "popUpWidnow",
    "Iframe",
    "age_of_domain",
    "DNSRecord",
    "web_traffic",
    "Page_Rank",
    "Google_Index",
    "Links_pointing_to_page",
    "Statistical_report",
    "Result"
]
print(dataset_filename)
print(f"{len(columns)}")
```

```python
import pandas as pd
import numpy as np
from scipy.io.arff import loadarff 

byte_data, metadata = loadarff(dataset_filename)
data = []
for index, row in enumerate(byte_data):
    data.append([int(cell.decode()) for cell in row])
data = np.array(np.array(data))
data = pd.DataFrame(data, columns=columns)

print(data)
```

```python
# Remove outliers
from scipy import stats

z_scores = np.abs(stats.zscore(data.select_dtypes(include=np.number)))
threshold = 2.5
outlier_mask = (z_scores > threshold).any(axis=1)

data_outlied = data[~outlier_mask]

print(f"{data.shape=}")
print(f"{data_outlied.shape=}")
```

```python
X = data_outlied.values
Y = data_outlied['Result']
X.shape, Y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
X_rescaled_minmax = scaler.fit_transform(X)

print(X)
print(X_rescaled_minmax)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_rescaled_standard = scaler.fit_transform(X)

print(X)
print(X_rescaled_standard)
```

## Feature selection

Statistical analysis (model-agnostic)

```python
from sklearn.feature_selection import SelectKBest, chi2 # chi2 can be non-sufficient/non-compatible to some data

feature_selector = SelectKBest(k=5, score_func=chi2)
test = feature_selector.fit(X_rescaled_minmax, Y)

print(test.scores_) # highest score == best

features = test.transform(X_rescaled_minmax)
print(X_rescaled_minmax[0: 3, :])
print(features[0: 3, :])
```

RFE - recursive feature elimination (model-based)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000) # Max iter so it doesnt complain about failing to converge
rfe = RFE(model, n_features_to_select=10)
test = rfe.fit(X, Y)

print(test.support_)
print(test.ranking_)

features = test.transform(X)
print(X[:, :])
print(features[:, :])
```

```python
test.support_
unimportant_columns = data_outlied.columns[~test.support_]
new_data = data_outlied.drop(columns=unimportant_columns)
new_data
```

# Model building

```python
from sklearn.model_selection import train_test_split
from sklearn import svm

X = new_data.values
Y = new_data['Result']
X.shape, Y.shape

print(X.shape, Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
# SVM
clf = svm.SVC(kernel='linear', C=1).fit(X_train, Y_train)
clf.score(X_test, Y_test)
```

```python
from sklearn.model_selection import cross_val_score
clf = svm.SVC(kernel='poly', C=1, random_state=42)
scores = cross_val_score(clf, X, Y, cv=10)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn import metrics
scores = cross_val_score(clf, X, Y, cv=5, scoring='f1_macro')
scores
```

```python
from sklearn.model_selection import ShuffleSplit
n_samples = X.shape[0]
cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
cross_val_score(clf, X, Y, cv=cv)
```
