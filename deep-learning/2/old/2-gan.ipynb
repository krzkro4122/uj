```python
# Funkcja generujca dane prawdziwe
def generate_real_samples(n, dim, a, b, c):
    x = np.linspace(-1, 1, n)
    y = a * x**2 + b * x + c + 0.1 * np.random.normal(size=n)
    x = x.reshape(n, 1)
    y = y.reshape(n, 1)
    labels = np.ones((n, 1))
    return np.hstack((x, y)), labels

# Funkcja generujca szumowe dane do oszukania dyskryminatora
def generate_fake_samples(n, dim):
    x = np.random.rand(n, dim) * 2 - 1
    y = np.random.rand(n, dim) * 2 - 1
    labels = np.zeros((n, 1))
    return np.hstack((x, y)), labels

# Funkcja budujca generator
def build_generator(latent_dim):
    model = models.Sequential()
    model.add(layers.Dense(10, input_dim=latent_dim, activation='relu'))
    model.add(layers.Dense(2, activation='linear'))
    return model

# Funkcja budujca dyskryminator
def build_discriminator(input_dim):
    model = models.Sequential()
    model.add(layers.Dense(25, input_dim=input_dim, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    return model

# Funkcja budujca model GAN
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = models.Sequential()
    model.add(generator)
    model.add(discriminator)
    return model

# Funkcja trenujca GAN
def train_gan(generator, discriminator, gan, epochs, batch_size, real_data_params):
    for epoch in range(epochs):
        # Generowanie danych prawdziwych
        real_data, real_labels = generate_real_samples(batch_size, 2, *real_data_params)

        # Generowanie szumnych danych do oszukania dyskryminatora
        fake_data, fake_labels = generate_fake_samples(batch_size, 1)

        # Trenowanie dyskryminatora na danych prawdziwych
        d_loss_real = discriminator.train_on_batch(real_data, real_labels)

        # Trenowanie dyskryminatora na danych wygenerowanych przez generator
        d_loss_fake = discriminator.train_on_batch(fake_data, fake_labels)

        # Obliczanie cakowitej straty dyskryminatora
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Generowanie losowego szumu dla GAN
        noise = np.random.rand(batch_size, latent_dim)

        # Tworzenie etykiet dla danych wygenerowanych przez generator (udajcych prawdziwe)
        valid_labels = np.ones((batch_size, 1))

        # Trenowanie GAN poprzez aktualizacj wag generatora przy zamroonym dyskryminatorze
        g_loss = gan.train_on_batch(noise, valid_labels)

        # Wywietlanie postpu co 100 epok
        if epoch % 100 == 0:
            print(f"Epoka {epoch}, Strata dyskryminatora: {d_loss[0]}, Strata generatora: {g_loss}")

            # Wizualizacja generowanych danych
            visualize_training(generator, real_data_params)

# Funkcja do wizualizacji generowanych danych w trakcie treningu
def visualize_training(generator, real_data_params):
    real_data, _ = generate_real_samples(100, 2, *real_data_params)
    fake_data, _ = generate_fake_samples(100, 2)
    generated_data = generator.predict(np.random.rand(100, latent_dim))

    plt.scatter(real_data[:, 0], real_data[:, 1], color='blue', label='Prawdziwe dane')
    plt.scatter(fake_data[:, 0], fake_data[:, 1], color='red', label='Generowane dane (faszywe)')
    plt.scatter(generated_data[:, 0], generated_data[:, 1], color='green', label='Generowane dane (prawdziwe)')
    plt.title('Wizualizacja generowanych danych')
    plt.legend()
    plt.show()

# Wymiary wejcia generatora
latent_dim = 10

# Wymiary danych wejciowych dla modeli
input_dim = 2

# Tworzenie generatora, dyskryminatora i modelu GAN
generator = build_generator(latent_dim)
discriminator = build_discriminator(input_dim)
gan = build_gan(generator, discriminator)

# Kompilacja modeli
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
gan.compile(loss='binary_crossentropy', optimizer='adam')

# Parametry danych rzeczywistych (funkcji wielomianowej)
real_data_params = (1, -2, 1)

# Trenowanie GAN
train_gan(generator, discriminator, gan, epochs=5000, batch_size=64, real_data_params=real_data_params)
```
