### Rotation

```python
ImageOps.flip(scaled_image)
```

### Cropping

```python
left, top, right, bottom = 0, 60, 0, 0
ImageOps.crop(
    scaled_image,
    (left, top, right, bottom)
)
```

```python
left, top, right, bottom = 20, 20, 30, 50
ImageOps.crop(
    scaled_image,
    (left, top, right, bottom)
)
```

### Centering & standarisation

```python
images = [
    Image.open("./images/rubble.jpg"),
    Image.open("./images/town.jpg"),
]
processed_images: list[Image.Image] = []
size = (128, 128)

for img in images:
    # scaled_img = img.copy()
    # scaled_img.thumbnail(size)
    # scaled_img = ImageOps.contain(img, size)
    scaled_img = ImageOps.fit(img, size)
    # scaled_img = ImageOps.pad(img, size, color="#fff")
    processed_images.append(scaled_img)
    print(img.size)

for img in processed_images:
    print(img.size)
    display(img)
```

### Standarisation

```python
import numpy
from PIL import Image, ImageFilter

standardised_images = []

for img in processed_images:
    # convert('RGB') for PNG file type
    converted_img = img.convert('RGB')
    pixels = numpy.asarray(converted_img)
    pixels = pixels.astype('float32')
    pixels /= 255.0
    standardised_images.append(Image.fromarray(pixels))
    # standardised_images.append(img.filter(ImageFilter.FIND_EDGES))

for img in standardised_images:    
    print(img.size)
    display(img)
```
