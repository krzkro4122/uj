# Image management

```python
from PIL import Image, ImageOps
```

### Loading an image

```python
image = Image.open("images/town.jpg")
```

### Displaying an image

```python
image
# image.show()
```

### Save to file

```python
image.save("images/output/town_saved.jpg")
```

### Basic image information

```python
image.format, image.size, image.mode
```

### Size changing

```python
size = (200, 300)

scaled_image = image.copy()
scaled_image.thumbnail(size)
# scaled_image = ImageOps.contain(image, size)
# scaled_image = ImageOps.fit(image, size)
# scaled_image = ImageOps.pad(image, size, color="#fff")

scaled_image.save("images/output/scaled_image.jpg")
scaled_image
```

### Rotation

```python
scaled_image.rotate(45)
```

### Fliping

```python
ImageOps.flip(scaled_image)
```

```python
ImageOps.mirror(scaled_image)
```

### Cropping

```python
left, top, right, bottom = 20, 50, 20, 50
ImageOps.crop(
    scaled_image,
    (left, top, right, bottom)
)
```

### Rescaling

```python
size = (200, 300)
scaled_image.resize(size)
```
